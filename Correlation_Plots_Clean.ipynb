```python
import seaborn as sns
import matplotlib.pyplot as plt
import os
import numpy as np

dir="Signal/" #<--- Insert signal-array folder
filename="stop_01_p_p_to_t1_t1~0_26.7494500003_39HT1001l2j.npy" #<--- Insert signal array filename
sig0=np.load(dir+filename,allow_pickle=True)

dir="Background/" #<--- Insert bg-array folder
filename="smHT1001l2j.npy" #<--- Insert bg array folder
bg0=np.load(dir+filename,allow_pickle=True)
```

```python
sigT=["HT","ETm","ETmphi","pt1","eta1","phi1","pt2","eta2","phi2","ptl","etal","phil"] #<--- Insert relevant parameters
sigU=["HT[MeV]","ETm[MeV]","ETmphi","pt1[MeV]","eta1","phi1","pt2[MeV]","eta2","phi2","ptl[MeV]","etal","phil"] #<--- Insert relevant parameters and units
```

```python
#The current implementation might require you to have a lot of ram available, I am not sure.
part=4
linesz=int(np.shape(sig0)[1]/part)
fig, axes = plt.subplots(nrows=np.shape(sig0)[1]*part, ncols=int(np.shape(sig0)[1]/part))
fig.set_figheight(300)
fig.set_figwidth(18)

for i in range(np.shape(sig0)[1]):
    for j in range(np.shape(sig0)[1]):
        if j==0:
            print(i)
        fig=plt.figure(figsize=(20,10))
        switches=int(j/linesz)
        axj=j-linesz*int(j/linesz)
        axi=part*i+int(j/linesz)
        sns.regplot(ax=axes[axi,axj], x=sig0[:300,j], y=sig0[:300,i], marker="+",label="Signal", truncate=False)
        sns.regplot(ax=axes[axi,axj], x=bg0[:300,j], y=bg0[:300,i], marker="+",label="Background", truncate=False)
        if axj==0:
            axes[axi,axj].set_ylabel(sigU[i], fontsize='xx-large')
        axes[axi,axj].set_xlabel(sigU[j], fontsize='xx-large')
```
